## Problem

$$
\lim_{x\to -1^-}\frac{3}{x+1}
$$

## Solution

As x approaches -1 from the left, x+1 approaches 0 from the negative side.

$$
\lim_{x\to -1^-}\frac{3}{x+1}=-\infty
$$

```python
from sympy import *

x = symbols('x')

expr = 3/(x + 1)

limit(expr, x, -1, dir='-')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_left = np.linspace(-5, -1.01, 500)
x_right = np.linspace(-0.99, 5, 500)

y_left = 3/(x_left + 1)
y_right = 3/(x_right + 1)

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x_left, y_left, label=r"$y=\frac{3}{x+1}$")
ax.plot(x_right, y_right)

# Center the axes
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Vertical asymptote
ax.axvline(x=-1, linestyle='--')

ax.grid(True)

plt.title(r"$\lim_{x\to -1^-}\frac{3}{x+1}$")
plt.show()
```
