# Limits with Python

## Exercise 1

Find:

$$
\lim_{x \to 3^+}\frac{1}{x-3}
$$

### Mathematical Explanation

As \(x\) approaches 3 from the right, \(x-3\) becomes a very small positive number.

Therefore,

$$
\lim_{x \to 3^+}\frac{1}{x-3}=+\infty
$$

```python
from sympy import *

# Define the variable
x = symbols('x')

# Define the function
f = 1/(x - 3)

# Calculate the right-hand limit as x approaches 3
result = limit(f, x, 3, "+")

print(result)
```

```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')

f = 1/(x - 3)

plot(f, (x, 0, 6))
```
